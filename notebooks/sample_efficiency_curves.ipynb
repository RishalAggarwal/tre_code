```python
import matplotlib
import numpy as np
import os

from matplotlib import pyplot as plt

plt.rcParams['text.usetex'] = True
plt.rcParams['text.latex.preamble'] = [r'\usepackage{sfmath} \boldmath']
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
%matplotlib inline
```

```python
load_path = ''  # set this

name = "logistic"  # loss function type
one_ratio_errors = np.load("{}/one_ratio_theta_diffs.npz".format(load_path))["theta_diffs"]
tre_errors = np.load("{}/tre_theta_diffs.npz".format(load_path))["theta_diffs"]
```

```python
tre_errors
```

```python
x_axis = np.array([10, 100, 1000, 10000, 100000])
fig, ax = plt.subplots(1, 1, figsize=(6, 6))

ax.plot(x_axis, one_ratio_errors, label="single ratio", c='b')
ax.plot(x_axis, tre_errors, label="TRE", c='r')

ax.set_xscale('log')
ax.tick_params(axis='x', labelsize=18)
ax.tick_params(axis='y', labelsize=18)
ax.set_ylabel(r'$\bf{| \theta* - \theta_{est}|}$', fontsize=28)
ax.set_xlabel(r"sample size (log scale)", fontsize=26)
ax.legend(fontsize=24)

fig.savefig(os.path.join("results/{}/{}_sample_efficiency.pdf".format(name, name)), dpi=300)
```
